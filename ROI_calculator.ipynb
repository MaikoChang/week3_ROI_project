```python
from IPython.display import clear_output

class Income:
    def __init__(self):
        self.rental_income = 0

class Expenses:    
    def __init__(self):
        self.tax = 0
        self.insurance = 0
        self.utilities = 0
        self.hoa_fee = 0
        self.lawn_snow = 0
        self.vacancy = 0
        self.repairs = 0
        self.capEx = 0
        self.prop_mgmt = 0
        self.mortgage = 0        
        self.total_expenses = 0
                       
class Cash_flow:
    def __init__(self):
        self.total_monthly_cashflow = 0
        self.total_annual_cashflow = self.total_monthly_cashflow*12
        
class Investment:
    def __init__(self):
        self.home_value = 0
        self.down_payment = 0
        self.closing_cost = 0
        self.rehab_budjet = 0
        self.misc_others = 0
        self.total_investment = 0
                        
class Overview:
    def __init__(self):
        self.income = Income()
        self.expenses = Expenses()
        self.cash_flow = Cash_flow()
        self.investment = Investment()
        
    def addIncome(self):
        self.rental_income = int(input("Enter your monthly rental income:"))
             
    def addExpenses(self):
        self.tax = int(input("Enter your monthly property tax:"))
        self.insurance = int(input("Enter your monthly home insurance payment:"))
        self.utilities = int(input("Enter your utilities per month if you pay. or enter '0' if you don't pay:"))
        self.hoa_fee = int(input("Enter your monthly HOA if you pay. or enter '0' if you don't pay:"))
        self.lawn_snow = int(input("Enter your cost for lawm/snow per month if you pay. or enter '0' if you don't pay:"))
        self.vacancy = int(input("Enter your vacancy rate without '%' :"))/100*self.rental_income
        self.repairs = int(input("Enter repair fee per month:"))
        self.capEx = int(input("Enter your capital expenditure per month:"))
        self.prop_mgmt = int(input("Enter your monthly property management fee:"))
        self.mortgage = int(input("Enter your monthly mortgage payment:"))
        self.total_expenses = self.tax + self.insurance + self.utilities + self.hoa_fee + self.lawn_snow + self.repairs + self.capEx + self.prop_mgmt + self.mortgage 
        
    def addCash_flow(self):
        self.total_monthly_cashflow = self.rental_income - self.total_expenses
        self.total_annual_cashflow = self.total_monthly_cashflow*12
    
    def addInvestment(self):
        self.home_value = int(input("Enter your Home value:"))
        self.down_payment = int(input("Enter your down payment:"))
        self.closing_cost = self.home_value*0.03
        self.rehab_budjet = int(input("Enter your budget for repair money:"))
        self.misc_others = int(input("Enter miscellaneous expenses:"))
        self.total_investment = self.down_payment + self.closing_cost + self.rehab_budjet + self.misc_others
        
    def cal_roi(self):
        print(f"Your annual return on investment for this property is: {self.total_annual_cashflow/self.total_investment*100}%")
                                  
    def run(self):
        self.addIncome()
        self.addExpenses()
        self.addCash_flow()
        self.addInvestment()
        self.cal_roi()
        
myhouse = Overview()
myhouse.run()
```
